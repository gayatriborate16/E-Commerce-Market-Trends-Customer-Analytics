```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
# replaceing missing values
df.replace(["?","Unknown","unknown","not_available",""], np.nan, inplace = True)
```

```python
# Remove Duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
# clean gender column
df["gender"] = df["gender"].replace({
    "M":"Male",
    "F":"Female",
    "male":"Male",
    "female":"Female"
})
df["gender"] = df["gender"].fillna(df["gender"].mode()[0])
```

```python
df["gender"]
```

```python
# clean city column
df["city"] = df["city"].str.strip()
df["city"] = df["city"].fillna(df["city"].mode()[0])
```

```python
df["city"]
```

```python
# clean age column
df["age"] = pd.to_numeric(df["age"], errors = "coerce")
```

```python
df["age"]
```

```python
df["age"] = df["age"].fillna(df["age"].median())
```

```python
df["age"] = df["age"].astype(int)
df["age"]
```

```python
# clean price column

df["price"] = df["price"].astype(str)

df["price"] = df["price"].str.replace(",", "")

df["price"] = df["price"].str.replace("free", "0")

df["price"] = pd.to_numeric(df["price"], errors="coerce")

df["price"] = df["price"].fillna(df["price"].mean())
```

```python
df["price"]
```

```python
# clean quantity column
df["quantity"] = df["quantity"].replace({
    "two":2,
    "three":3
})
df["quantity"] = df["quantity"].fillna(df["quantity"].mode()[0])
```

```python
df["quantity"] = df["quantity"].astype(int)
```

```python
df["quantity"]
```

```python
# clean discount column
df["discount_percent"] = df["discount_percent"].astype(str)
df["discount_percent"] = df["discount_percent"].replace("%","")
df["discount_percent"] = pd.to_numeric(df["discount_percent"], errors = "coerce")
df["discount_percent"] = df["discount_percent"].fillna(0)
```

```python
df["discount_percent"]
```

```python
df["discount_percent"] = df["discount_percent"].astype(int)
```

```python
df["discount_percent"]
```

```python
# Clean Date Column
df["order_date"] = pd.to_datetime(df["order_date"], errors = "coerce")
df["order_date"] = df['order_date'].fillna(method='ffill')
```

```python
df["order_date"]
```

```python
# Clean Payment Method column
df["payment_method"] = df["payment_method"].replace({
    "upi":"UPI",
    "card":"Card"
})
df["payment_method"] = df["payment_method"].fillna(df["payment_method"].mode()[0])
```

```python
df["payment_method"]
```

```python
# clean customer rating column
df["customer_rating"] = df["customer_rating"].replace({
    "five":5,
    "?":""
})
df["customer_rating"] = pd.to_numeric(df["customer_rating"], errors = "coerce")
```

```python
df["customer_rating"] = df["customer_rating"].fillna(df["customer_rating"].mean())
```

```python
df["customer_rating"] = df["customer_rating"].astype(int)
```

```python
df["customer_rating"]
```

```python
# Clean Delivery Days column
df["delivery_days"] = df["delivery_days"].replace({
    "?":"",
    "unknown":""
})
```

```python
df["delivery_days"]
```

```python
df["delivery_days"] = pd.to_numeric(df["delivery_days"], errors = "coerce")
df["delivery_days"] = df["delivery_days"].fillna(df["delivery_days"].mean())
```

```python
df["delivery_days"] = df["delivery_days"].astype(int)
```

```python
df["delivery_days"]
```

```python
# Clean Returned Column
df["returned"] = df["returned"].replace({
    "Y":"Yes",
    "N":"No"
})
df["returned"] = df["returned"].fillna(df["returned"].mode()[0])
```

```python
df["returned"]
```

```python
# Handle Outliers (IQR Method)
numeric_cols = df.select_dtypes(include = ["int64","float64"]).columns
```

```python
for cols in numeric_cols:
    Q1 = df[cols].quantile(0.25)
    Q3 = df[cols].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[cols] = np.where(df[cols] < lower, lower, df[cols])
    df[cols] = np.where(df[cols] > upper, upper, df[cols])
```
